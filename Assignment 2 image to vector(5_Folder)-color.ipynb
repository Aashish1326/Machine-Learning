```python
vd1=cv2.VideoCapture(0)
count=0
while True:
    count=count+1
    su,img=vd1.read()
    cv2.imshow("video",img)
    cv2.imwrite(r"D:\Data Sets\OpenCv5folder\4\{}.jpg". format(count),img)
    if cv2.waitKey(1000)==ord("c"):
        break
cv2.destroyAllWindows()
```

##### `imread`--->loads image and returns it 

##### `imshow`--->displays an image

##### `imwrite`--->saves images to specified directory

#### `waitkey`--->allows users to display a window for given time untill any key is pressed
